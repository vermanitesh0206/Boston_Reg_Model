```python
import pandas as pd
```

```python
df=pd.read_csv('./boston.csv')
```

```python
df.describe()
```

```python
def normal(d):
    p=(d-d.mean())/d.std()
    return p
```

```python
df['CRIM1']=normal(df['CRIM'])
df['ZN1']=normal(df['ZN'])
df['INDUS1']=normal(df['INDUS'])
df['CHAS1']=normal(df['CHAS'])
df['NOX1']=normal(df['NOX'])
df['RM1']=normal(df['RM'])
df['AGE1']=normal(df['AGE'])
df['DIS1']=normal(df['DIS'])
df['RAD1']=normal(df['RAD'])
df['TAX1']=normal(df['TAX'])
df['PTRATIO1']=normal(df['PTRATIO'])
df['B1']=normal(df['B'])
df['LSTAT1']=normal(df['LSTAT'])
```

```python
df1 = df.drop(['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV'],axis=1)
```

```python
df1.dtypes
```

```python
dm_m = df1#.drop(['ZN1'],axis=1)
```

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test = train_test_split(dm_m, df.MEDV, test_size=0.4, random_state=140)
```

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm
```

```python
lm.fit(x_train,y_train)
```

```python
y_pred = lm.predict(x_test)
```

```python
df1 = pd.DataFrame(x_train.columns,lm.coef_ )
print(df1.shape)
print(df1)
```

```python
lm.score(x_test,y_test)
```

```python
df2 = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
df2
```
